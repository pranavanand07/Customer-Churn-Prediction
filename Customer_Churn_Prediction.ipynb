Feature Engineering and Model Preparation

```python
# seperating target variable (y) from feature variable (X)
X=df.drop(['Churn','customerID'],axis=1)
X
```

```python
y=df['Churn']
y
```

```python
categorical_cols=X.select_dtypes(include=['object','category']).columns
```

```python
numerical_cols=X.select_dtypes(include=['int64','float64']).columns
```

```python
print("categorical columns :",categorical_cols)
print("Numerical columns :",numerical_cols)
```

```python
#converting categorical data into numerical ones using one hot encoding
from sklearn.preprocessing import OneHotEncoder
```

```python
X_encoded=pd.get_dummies(X,columns=categorical_cols,drop_first=True)
```

```python
len(X_encoded.columns)
```

```python
len(X.columns)
```

```python
X_encoded.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
#split data into training and testing
X_train,X_test,y_train,y_test=train_test_split(
    X_encoded,y,test_size=0.2,random_state=42
    )
```

```python
print("Shape of X_train:",X_train.shape)
print("Shape of X_test:",X_test.shape)
```

```python
print("Shape of y_train:",y_train.shape)
print("Shape of y_test:",y_test.shape)
```

```python
## Scaling numerical values
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
X_train[numerical_cols]=scaler.fit_transform(X_train[numerical_cols])
X_test[numerical_cols]=scaler.fit_transform(X_test[numerical_cols])
```

```python
X_train.head()
```

Building and Training Model

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

Logistic Regression

```python
model=LogisticRegression(max_iter=1000)
```

```python
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
```

Model Performance

```python
accuracy=accuracy_score(y_test,y_pred)
print("Accuracy:",accuracy)
```

```python
cm=confusion_matrix(y_test,y_pred)
print("Confusion Matrix:")
print(cm)
```

```python
plt.figure(figsize=(6, 4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Did Not Churn (0)', 'Churned (1)'],
            yticklabels=['Did Not Churn (0)', 'Churned (1)'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

```python
print(classification_report(y_test, y_pred, target_names=['Did Not Churn (0)', 'Churned (1)']))
```

Random Forest

```python
rf=RandomForestClassifier(n_estimators=100,random_state=42,n_jobs=-1)
```

```python
rf=model.fit(X_train,y_train)
```

Handling imbalance

```python
print("--- Training Logistic Regression (Weighted) ---")
# The *only* change is adding class_weight='balanced'
lr_weighted_model = LogisticRegression(max_iter=1000, class_weight='balanced')
lr_weighted_model.fit(X_train, y_train)
```

```python
y_pred_lr_weighted = lr_weighted_model.predict(X_test)
```

```python
print(classification_report(y_test, y_pred_lr_weighted, target_names=['Did Not Churn (0)', 'Churned (1)']))
```

```python
print("\n\n--- Training Random Forest (Weighted) ---")
# Again, the *only* change is adding class_weight='balanced'
rf_weighted_model = RandomForestClassifier(n_estimators=100,
                                           random_state=42,
                                           n_jobs=-1,
                                           class_weight='balanced')
rf_weighted_model.fit(X_train, y_train)
```

```python
y_pred_rf_weighted = rf_weighted_model.predict(X_test)
```

```python
print(classification_report(y_test, y_pred_rf_weighted, target_names=['Did Not Churn (0)', 'Churned (1)']))
```

```python
# --- 1. Get the Coefficients ---
# 'coef_[0]' gets the list of coefficients for our one target class
coefficients = lr_weighted_model.coef_[0]

# Get the feature names from our training data
feature_names = X_train.columns

# --- 2. Create a DataFrame for easy viewing ---
# We use 'np.exp()' to make the numbers easier to interpret.
# This converts them from "log-odds" to "odds ratios".
# An odds ratio > 1 increases churn. An odds ratio < 1 decreases churn.
odds_ratios = np.exp(coefficients)

coef_df = pd.DataFrame({
    'Feature': feature_names,
    'Coefficient': coefficients,
    'Odds Ratio': odds_ratios
})

# Sort by the *absolute* value to see the most impactful features (both good and bad)
coef_df['Impact_Magnitude'] = coef_df['Coefficient'].abs()
coef_df = coef_df.sort_values(by='Impact_Magnitude', ascending=False)

print("--- Most Impactful Features (from Weighted LR Model) ---")
print(coef_df.head(15))


# --- 3. Plot the Top 10 Features ---
# We'll plot the 'Odds Ratio' as it's more intuitive

# Separate into positive (increases churn) and negative (decreases churn)
top_features = coef_df.head(15).sort_values(by='Odds Ratio', ascending=False)

plt.figure(figsize=(10, 8))
sns.barplot(x='Odds Ratio', y='Feature', data=top_features, palette='RdBu_r')
plt.axvline(x=1, color='black', linestyle='--') # Add a line at 1.0
plt.title('Top 15 Features Influencing Churn (Odds Ratio)')
plt.xlabel('Odds Ratio ( > 1 Increases Churn, < 1 Decreases Churn)')
plt.show()
```
